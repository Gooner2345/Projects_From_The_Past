<h4> 4.5) Strategy for Imputing the missing value:</h4>
Lets find the suitable imputing strategy<br>
As we have at least one missing value in each column we cannot use dropna because that will delete all values

<Strong>Our Strategy:</Strong> 
- If a column has outliers we will use median to fill the NAs as median value is not effected by the outliers and by doing this we will not effect the distribution of the curve as well. (Only for columns that has 2 to 5 missing values)<br>
<br>
- For the cases with out any outliers we will use mean for imputing (Only for columns that has 2 to 5 missing values)<br>
<br>
- If a column has more than 5 missing values, we will look into them critically to understand relation between missing variables and other variables across countries of the same region. It is important to mention that cross-country comparisons will be done based on trends/patterns observed<br>
<br>
- Along with the strategy mentioned above, we will use literature review to see if our findings are supported by available secondary data<br>
